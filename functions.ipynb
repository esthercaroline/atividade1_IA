```python
from aigyminsper.search.SearchAlgorithms import BuscaLargura
from aigyminsper.search.SearchAlgorithms import BuscaProfundidade
from aigyminsper.search.SearchAlgorithms import BuscaProfundidadeIterativa
from aigyminsper.search.Graph import State
from datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt
```

```python
class SumOne(State):

    def __init__(self, n, op, g):
        self.operator = op
        self.number = n
        self.goal = g

    def successors(self):
        sucessors = []
        if self.number < self.goal:
            sucessors.append(SumOne(self.number+1, "+1 ", self.goal))
            sucessors.append(SumOne(self.number+2, "+2 ", self.goal))
        return sucessors

    def is_goal(self):
        if self.goal == self.number:
            return True
        return False

    def description(self):
        return "Este é um agente simples que sabe somar 1 e 2"

    def cost(self):
        return 1

    def env(self):
        return self.number
```

```python
def busca_profundidade_10(objetivo):
    start = datetime.now()
    b = SumOne(0, "", objetivo)
    bp = BuscaProfundidade()
    result = bp.search(b,10)
    end = datetime.now() 
    if result != None:
        return end - start
    else:
        return "Não foi possível encontrar a solução"

dic_tempos_profundidade10 = {
    'Iteração': [],
    'Tempo': []
}
for i in range(50):
    tempo = busca_profundidade_10(i)
    dic_tempos_profundidade10['Iteração'].append(i)
    dic_tempos_profundidade10['Tempo'].append(tempo)
print(dic_tempos_profundidade10)
```

```python
# Gráfico

# Função para converter o tempo e lidar com a string "Não foi possível encontrar a solução"
def convert_time(x):
    if x == 'Não foi possível encontrar a solução':
        return -10
    else:
        return float(x.split(':')[-1])

# Carregar os DataFrames
dfbp10 = pd.read_csv("bp10.csv")
dfbp100 = pd.read_csv("bp100.csv")
dfbpi = pd.read_csv("bpi.csv")
dfbl = pd.read_csv("bl.csv")

plt.figure(figsize=(8, 6))


dfbp10['Segundos'] = dfbp10['Tempo'].apply(convert_time)
dfbp100['Segundos'] = dfbp100['Tempo'].apply(convert_time)
dfbpi['Segundos'] = dfbpi['Tempo'].apply(convert_time)
dfbl['Segundos'] = dfbl['Tempo'].apply(convert_time)

# Plotar Busca em Profundidade 10
plt.plot(dfbp10['Iteração'], dfbp10['Segundos'], label='Busca em Profundidade 10')

# Plotar Busca em Profundidade 100
plt.plot(dfbp100['Iteração'], dfbp100['Segundos'], label='Busca em Profundidade 100')

# Plotar Busca em Profundidade Iterativa
plt.plot(dfbpi['Iteração'], dfbpi['Segundos'], label='Busca em Profundidade Iterativa')

# Plotar Busca em Largura
plt.plot(dfbl['Iteração'], dfbl['Segundos'], label='Busca em Largura')

plt.xlabel('Iteração')
plt.ylabel('Tempo (s)')
plt.title('Tempos de Busca com objetivos de 0 a 49')
plt.legend()
plt.grid(True)

plt.xticks(range(0, 51, 5))

plt.show()
```
